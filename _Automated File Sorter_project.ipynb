# Automated File Sorter In File Explorer

```python
import os, shutil             # importing necessary package 
```

```python
path = r"C:/Users/91994/Documents/Python_tutorials/"
```

```python
file_name = os.listdir(path)     # list of the files in the folder(path)
```

```python
folder_names = ["ipynb files", "png files", "pdf files"]      # list of file types
for loop in range(0,3):
    if not os.path.exists(path + folder_names[loop]):
        os.makedirs(path + folder_names[loop])
        
```

```python
for file in file_name:                                                       # sorting
    if ".pdf" in file and not os.path.exists(path + "pdf files/" + file):       #  for pdf files
        shutil.move(path + file, path + "pdf files/" + file)
    elif ".ipynb" in file and not os.path.exists(path + "ipynb files/" + file):   # for ipynb files
        shutil.move(path + file, path + "ipynb files/" + file)
    elif ".png" in file and not os.path.exists(path + "png files/" + file):       # for png files
        shutil.move(path + file, path + "png files/" + file)
    else:
        print("There are files in the path that were not moved")
```
