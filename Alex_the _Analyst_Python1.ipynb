```python
daily_log1 = {1,3,52,4,5,3,34,54}
daily_log2 = {32,4,54,56,87,34}

print(daily_log1 | daily_log2)   # the unique values present both , either in log1 or log2

print(daily_log1 & daily_log2)   # the unique values present in both 1 and 2

print(daily_log1 - daily_log2)   # the values not in the log1

print(daily_log2 - daily_log1)   # the values only in log2

print(daily_log1 ^ daily_log2)   # the unique values only present in either 1 or 2
```

## Functions

```python
def first_function():     # without arguements
    print("we did it !")

first_function()


def num_squared(num):                #  with arguements
    print(num**2)
print('num_squared')
num_squared(5)


def num_squared_cus(num, power):      # with 2 arguements
    print(num**power)
print("num_squared_cus")
num_squared_cus(5,3)                  # if we have 2 arguements, then we need to pass 2 arguements when calling

num_squared_cus(power = 5, num = 3)    # custom



args_tuple = (5,6,1,2,8)
def num_args(* number):                  # when we don't know the no.of arguements we need to pass
    print(number[0]*number[1])
print("num_args")
num_args(*args_tuple)                     # don't forget to put *


# with keyword arguement
print("with keyword arguement")
def num_kwargs(**num):             # for arbituary keyword arguement " ** " in the def()
    print("My number is " + num["integer"]  + ". My other number is "+ num["integer1"])

num_kwargs(integer = "500", integer1 = "505")
```

# BMI calculator - project

```python
name = input("Enter your name = ")
weight  = float(input(" weight (in Kgs) = "))
height  = float(input(" height (in meters) = "))

BMI = (weight)/ (height ** 2)
print("Your BMI = ",BMI)

if BMI > 0:
    if BMI <= 18.5:
        print(name + ", You are underweight")
    elif BMI <= 24.9:
        print(name + ", You are normal weight")
    elif BMI <= 29.9:
        print(name + ", You are overweight. You need to more movement")
    elif BMI >= 30:
        print(name + ", You are obese. You need to control your weight inorder to avoid severe helath problems")
else:
    print("Invalid input!")
```
